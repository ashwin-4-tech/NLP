```python
import re

def preprocess_text(text):
    """ Transforms raw text into clean, structured tokens. """

    # 1. Convert text to lowercase
    text = text.lower()

    # 2. Remove URLs and email-like patterns
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Handle repeated characters (e.g., "soooo goooood!!!" -> "so good")
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Remove special characters/punctuation (optional but good for cleaning)
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. Remove very short tokens (length <= 2) with exceptions
    words = text.split()
    processed_words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    # 7. Remove extra spaces
    return " ".join(processed_words).strip()

# --- Stress Testing ---

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print(f"{'ORIGINAL TEXT':<40} | {'CLEANED TEXT'}")
print("-" * 80)
for original in sample_inputs:
    cleaned = preprocess_text(original)
    print(f"{original:<40} | {cleaned}")
```
